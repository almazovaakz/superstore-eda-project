```python
data = data.dropna(subset=['Postal Code'])
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True)
data['Ship Date'] = pd.to_datetime(data['Ship Date'], dayfirst=True)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x='Segment', y='Sales', estimator=sum)
plt.title('Total sales by customer segment')
plt.ylabel('Total sales')
plt.xlabel('Segment')
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x='Category', y='Sales', estimator=sum)
plt.title('Total sales by product category')
plt.ylabel('Total sales')
plt.xlabel('Category')
plt.show()
```

```python
data['Month'] = data['Order Date'].dt.to_period('M')

monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Sales')
plt.title('Sales dynamics by months')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Month'] = data['Order Date'].dt.to_period('M').dt.to_timestamp()

monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Sales')
plt.title('Sales dynamics by months')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()





```

```python
segment_sales = data.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
print(segment_sales)
```
